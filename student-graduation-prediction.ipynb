```python
file_path = r'C:\Users\HP\Documents\bilgisayar_ogrenci_verileri.xlsx'

df = pd.read_excel(file_path)
```

```python
career_path_column = df['Uzmanlık Alanı'].drop_duplicates().tolist()
print(career_path_column)
```

```python
grade_columns = df.columns[:-1]  

from sklearn.preprocessing import LabelEncoder

le_grades = {col: LabelEncoder() for col in grade_columns}

for col in grade_columns:
    if df[col].isnull().sum() == 0:  
        df[col] = le_grades[col].fit_transform(df[col])
    else:
        print(f"Column '{col}' contains null values. Please handle them before encoding.")
```

```python
df.head()
```

```python
career_path_column = "Uzmanlık Alanı"

le_career_path = LabelEncoder()

if career_path_column in df.columns:
    df[career_path_column] = le_career_path.fit_transform(df[career_path_column])
else:
    print(f"Column '{career_path_column}' does not exist in the DataFrame.")
```

```python
df.head()
```

```python
inputs = []
embeddings = []
embedding_size = 5  
```

```python
for col in grade_columns:
    input_layer = Input(shape=(1,), name=f'{col}_input')
    
    embedding_layer = Embedding(
        input_dim=len(le_grades[col].classes_),  
        output_dim=embedding_size,            
        name=f'{col}_embedding'
    )(input_layer)
    
    flat_layer = Flatten()(embedding_layer)
    
    inputs.append(input_layer)
    embeddings.append(flat_layer)
```

```python
x = concatenate(embeddings)

x = Dense(128, activation='relu')(x)
x = Dropout(0.5)(x)  
x = Dense(64, activation='relu')(x)

output = Dense(len(le_career_path.classes_), activation='softmax')(x)

model = Model(inputs=inputs, outputs=output)

model.summary()
```

```python
model = Model(inputs=inputs, outputs=output)
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```
